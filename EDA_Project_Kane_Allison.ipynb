Maintaining metabolic health is imperative to preventing the development of chronic health conditions like diabetes, heart disease, and more. In screening for metabolic health, patients with a higher Body Mass Index (BMI) are often labeled at risk for developing these conditions. However, BMI is highly criticized as a metric of overall patient health, failing to take into account muscle mass, body fat distribution, gender, genetics and diet. This could bias assessments of health to underdiagnose "lean" individuals, delaying critical testing and care. Understanding how to better diagnose these populations is critical, especially with the rise in prevalence of diabetes among patients with lower BMI (7).

Existing standards for determining overall health like BMI fail to capture the entire picture of how diet, exercise, genetics, and social determinants of health contribute to patient risk and overall outcomes. This analysis will focus on how measures of dietary fitness (dietary fiber, protein, carbohydrate, total sugar, and fat intake), demographic information (gender, poverty level, racial/ethnic background) and detailed medical assessment tools provide a clearer assessment of diabetes and prediabetes risk than BMI alone. Fasting glucose and diabetes diagnosis status will be key outcome indicator of subject metabolic health. The goal is to develop a better standard of risk assessment and increase patient awareness of diabetes risk at any BMI, incorporating all aspects contributing to patient health.

## INSPECTING DEMOGRAPHICS
In the case of the NHANES data, some categorical variables are recorded in a numeric form to consolidate the data collection process.
```python
# some initial cleaning
print(demographics['SEQN'].dtype,
demographics['RIAGENDR'].dtype,
demographics['RIDRETH3'].dtype,)

# converting number code to codebook value
demographics['RIAGENDR'] = demographics['RIAGENDR'].replace({1.0: 'male', 2.0: 'female'})
demographics['RIDRETH3'] = demographics['RIDRETH3'].replace({1.0 : 'Mexican American',
                                                             2.0 : 'Other Hispanic',
                                                             3.0 : 'Non-Hispanic White',
                                                             4.0 : 'Non-Hispanic Black',
                                                             6.0 : 'Non-Hispanic Asian',
                                                             7.0 : 'Other Race - Including Multi-Racial'})

# seqn is a unique value for each participant, 
demographics['SEQN'] = demographics['SEQN'].astype(str)
```
